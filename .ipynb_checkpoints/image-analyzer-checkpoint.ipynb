Keras API prefers data to be divided into separate train/ and test/ directories, and under each directory to have a subdirectory for each class, e.g. a train/dogs/ ... subdirectories and the same for test. Images are then organized under the subdirectories.

The train folder should contain ‘n’ folders each containing images of respective classes. For example, In the Dog or Cats data set, the train folder should have 2 folders, namely “dogs” and “cats” containing respective images inside them.

```python
# Create above directory structure
dataset_home = 'dataset/final_dataset/'
subdirs = ['train/', 'validation/']
for subdir in subdirs:
    # create label subdirectories
    labeldirs = ['cars/', 'cats/', 'dogs/', 'motorcycles/' ]
    for labldir in labeldirs:
        newdir = dataset_home + subdir + labldir
        makedirs(newdir, exist_ok=True)
```

<b>Pre-Process Photo Sizes & Reshape images to 200x200x3 pixels<b/>

```python
# Cars
src_directory = 'dataset/cars/'
for file in glob.iglob(src_directory + '**/*.jpg', recursive=True):
    im = Image.open(file)
    imResize = im.resize((200,200), Image.ANTIALIAS)
    imResize.save(file , 'JPEG', quality=90)
    
# Motorcycles
src_directory = 'dataset/motorcycle/'
for file in glob.iglob(src_directory + '**/*.jpg', recursive=True):
    im = Image.open(file)
    imResize = im.resize((200,200), Image.ANTIALIAS)
    imResize.save(file , 'JPEG', quality=90)
    
# Cats and Dogs (Because they are in same directory)
src_directory = 'dataset/cats_dogs/train/'
for file in glob.iglob(src_directory + '**/*.jpg', recursive=True):
    im = Image.open(file)
    imResize = im.resize((200,200), Image.ANTIALIAS)
    imResize.save(file , 'JPEG', quality=90)
```

<b>Split the datasets to train and test (75% Train, 25% Test) & Copy the datasets to corresponding subdirectories in dataset/final_dataset/. This is done consistently by fixing the seed for the pseudorandom number generator so that we get the same split of data each time the code is run.<b/>

```python
# seed random number generator
seed(1)
# define ratio of pictures to use for testing
test_ratio = 0.25

# Cars
src_directory = 'dataset/cars/'
for file in listdir(src_directory):
    src = src_directory + '/' + file
    dst_dir = 'train/'
    if random() < test_ratio:
        dst_dir = 'validation/'
    dst = dataset_home + dst_dir + 'cars/'  + file
    copyfile(src, dst)

# Motorcycles
src_directory = 'dataset/motorcycle/'
for file in listdir(src_directory):
    src = src_directory + '/' + file
    dst_dir = 'train/'
    if random() < test_ratio:
        dst_dir = 'validation/'
    dst = dataset_home + dst_dir + 'motorcycles/'  + file
    copyfile(src, dst)

# Cats and Dogs (Because they are in same directory)
src_directory = 'dataset/cats_dogs/train/'
for file in listdir(src_directory):
    src = src_directory + '/' + file
    dst_dir = 'train/'
    if random() < test_ratio:
        dst_dir = 'validation/'
    if file.startswith('cat'):
        dst = dataset_home + dst_dir + 'cats/'  + file
        copyfile(src, dst)
    elif file.startswith('dog'):
        dst = dataset_home + dst_dir + 'dogs/'  + file
        copyfile(src, dst)
```

## Final Data Preparation before modeling

Format the images into appropriately pre-processed floating point tensors before feeding to the network:

1. Read images from the disk.
2. Decode contents of these images and convert it into proper grid format as per their RGB content.
3. Convert them into floating point tensors.
4. Rescale the tensors from values between 0 and 255 to values between 0 and 1, as neural networks prefer to deal with small input values.

Fortunately, all these tasks can be done with the `ImageDataGenerator` class provided by `tf.keras`. It can read images from disk and preprocess them into proper tensors. It will also set up generators that convert these images into batches of tensors—helpful when training the network.

<b>ImageDataGenerator:<b/> Generate batches of tensor image data with real-time data augmentation. E.g. rescales the images

```python
train_image_generator = ImageDataGenerator(rescale=1./255) # Generator for our training data
test_image_generator = ImageDataGenerator(rescale=1./255) # Generator for our test data
```

After defining the generators for training and validation images, the `flow_from_directory` method load images from the disk, applies rescaling, and resizes the images into the required dimensions.

<b>flow_from_directory:<b/> Takes the path to a directory & generates batches of augmented data.
    
<b>class_mode:<b/> One of "categorical", "binary", "sparse", "input", or None. Default: "categorical". Determines the type of label arrays that are returned: - "categorical" will be 2D one-hot encoded labels, - "binary" will be 1D binary labels, "sparse" will be 1D integer labels, - "input" will be images identical to input images (mainly used to work with autoencoders). - If None, no labels are returned (the generator will only yield batches of image data, which is useful to use with model.predict_generator()). Please note that in case of class_mode None, the data still needs to reside in a subdirectory of directory for it to work correctly. Set “binary” if you have only two classes to predict, if not set to“categorical”, in case if you’re developing an Autoencoder system, both input and the output would probably be the same image, for this case set to “input”.
    
* The directory must be set to the path where your ‘n’ classes of folders are present. directory: string, path to the target directory. It should contain one subdirectory per class. Any PNG, JPG, BMP, PPM or TIF images inside each of the subdirectories directory tree will be included in the generator.
* The target_size is the size of your input images, every image will be resized to this size
* shuffle: Set True if you want to shuffle the order of the image that is being yielded, else set False.

# Modeling part 1 (CNN unoptimized)

For convenience, set up variables to use while pre-processing the dataset and training the network.

<b>batch_size:<b/>
    Defines the number of samples that will be propagated through the network. For instance, let's say you have 1050 training samples and you want to set up a batch_size equal to 100. The algorithm takes the first 100 samples (from 1st to 100th) from the training dataset and trains the network. Next, it takes the second 100 samples (from 101st to 200th) and trains the network again... It requires less memory. Since you train the network using fewer samples, the overall training procedure requires less memory. That's especially important if you are not able to fit the whole dataset in your machine's memory.
    
<b>epochs:<b/>
    The number of epochs is a hyperparameter that defines the number times that the learning algorithm will work through the entire training dataset. One epoch means that each sample in the training dataset has had an opportunity to update the internal model parameters. An epoch is comprised of one or more batches. You can think of a for-loop over the number of epochs where each loop proceeds over the training dataset. Within this for-loop is another nested for-loop that iterates over each batch of samples, where one batch has the specified “batch size” number of samples. The number of epochs is traditionally large, often hundreds or thousands, allowing the learning algorithm to run until the error from the model has been sufficiently minimized.
